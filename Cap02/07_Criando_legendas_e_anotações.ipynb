```python
import pandas as pd

import matplotlib.pyplot as plt
from matplotlib import rcParams
import seaborn as sns


%matplotlib inline
rcParams['figure.figsize'] = 8, 4
sns.set_style('whitegrid')

caminho = 'C:/Users/lucas/OneDrive/Documentos/GitHub/Python Ciência de Dados/dados/mtcars.csv'
carros = pd.read_csv(caminho)
carros.columns = ['nomes','mpg','cyl','disp', 'hpt', 'drat', 'wt', 'qsec', 'vs', 'am', 'gear', 'carb']
mpg = carros['mpg']
```

### Colocando títulos nos eixos

```python
x = range(1, 10)
y = [1, 2, 3, 4, .5, 4, 3, 2, 1]
plt.bar(x, y)
plt.xlabel('Nosso título do eixo X')
plt.ylabel('Nosso título do eixo Y')
```

### Colocalndo legenda no gráfico

```python
from logging import handlers
import matplotlib.patches as mpatches
from matplotlib.pyplot import legend

plt.plot(mpg)
plt.xlabel('Nome dos carros')
plt.ylabel('milhas/galão')
plt.title('Consumo de gasolina')
plt.xticks(mpg.index, carros.nomes, rotation=45)

legenda = mpatches.Patch(label='mpg')
plt.legend(handles=[legenda])
```
