```python
import numpy as np
import pandas as pd

from matplotlib import pyplot as plt
from matplotlib import rcParams
import seaborn as sns

%matplotlib inline
rcParams['figure.figsize'] = 5, 4
sns.set_style('whitegrid')
```

### Supondo a distribuição dos dados com histogramas

```python
caminho = 'C:/Users/lucas/OneDrive/Documentos/GitHub/Python Ciência de Dados/dados/mtcars.csv'
carros = pd.read_csv(caminho)
carros.columns = ['nomes','mpg','cyl','disp', 'hp', 'drat', 'wt', 'qsec', 'vs', 'am', 'gear', 'carb']
carros.index = carros.nomes
mpg = carros['mpg']

plt.hist(mpg)
plt.plot()
plt.show()
```

```python
sns.distplot(mpg)
```

###  Vendo os scatterplots em ação

```python
carros.plot(kind='scatter', x='hp', y='mpg', c=['#9A9A9A'], s=150)
```

```python
sns.regplot(x='hp', y='mpg', data=carros, scatter=True)
```

### Gerando uma matriz de scatter plots

```python
sns.pairplot(carros)
```

```python
colunas = ['mpg', 'disp', 'hp', 'wt']
novo_df = pd.DataFrame((carros[colunas].values), columns = colunas)
carros_target = carros['am'].values

novo_df['group'] = pd.Series(carros_target, dtype='category')
```

```python
sns.pairplot(novo_df, hue='group', palette='hls')
```

### Construindo boxplots

```python
carros.boxplot(column='mpg', by='am')
carros.boxplot(column='wt', by='am')
```

```python
sns.boxplot(x='am', y='mpg', data=carros, palette='hls')
```
